```python
import math
import statistics
import numpy as np
import scipy.stats
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
gdp_data = pd.read_csv(r'C:\Users\Patrick\Nextcloud\Studium\FHDW\Data Analysis and Machine learning\gdp_csv.csv')
internet_speed_data = pd.read_csv(r'C:\Users\Patrick\Nextcloud\Studium\FHDW\Data Analysis and Machine learning\internet broadband and mobile speeds by country.csv')
gdp = pd.DataFrame(gdp_data, columns=['Country Name', 'Country Code', 'Value', 'Year'])
internet_speed = pd.DataFrame(internet_speed_data, columns=['Country','Broadband Mbps', 'Mobile Mbps', 'As of'])
gdp.rename(columns = {'Country Code':'CountryCode', 'Country Name':'CountryName', 'Year':'YearGDP', 'Value':'GDP'}, inplace = True)
internet_speed.rename(columns = {'Country':'CountryName', 'Broadband Mbps':'BroadbandSpeed', 'Mobile Mbps':'MobileSpeed', 'As of':'YearInternetSpeed'}, inplace = True)
dataset = pd.merge(gdp, internet_speed, how='inner', on = 'CountryName')
dataset['GDP'] = dataset['GDP'].div(1000000000).round(2)
```

```python
plotBroadBand = dataset.plot.scatter(x="GDP", y="BroadbandSpeed")
plotBroadBand.set_xlabel('GDP in billion USD')
plotBroadBand.set_ylabel('Broadband Speed in Mbps')
plotBroadBand.set_xlim(0,20000)
```

```python
plotBroadBand = dataset.plot.scatter(x="GDP", y="MobileSpeed")
plotBroadBand.set_xlabel('GDP in billion USD')
plotBroadBand.set_ylabel('Mobile Speed in Mbps')
plotBroadBand.set_xlim(0, 20000)
```
